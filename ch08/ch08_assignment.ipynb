```python
a = [3, 6, 7, 4, 9, 10, 13]

def swap_num():
  for i in range(len(a)):
    if a[i] % 2 == 0:
      even_id = i
      break
  for i in range(len(a)):
    if a[i] % 2 != 0:
      odd_id = i

  temp = a[even_id]
  a[even_id] = a[odd_id]
  a[odd_id] = temp
  return a

swap_num()
```

```python
a = [3, 6, 7, 4, 9, 10, 13]

def find_max(lista):
  max_num = lista[0]
  max_index = 0
  for i in range(1, len(lista)):
    if max_num < lista[i]:
      max_num = lista[i]
      max_index = i

  print("최댓값:", max_num, "인덱스:", max_index)
  return max_num, max_index

find_max(a)
```

```python
dict_a = {'a' : 10, 'b' : 20, 'c' : 30}

def dict_sum(dict_a):
  sum = 0
  for i in dict_a.values():
    sum += i

  return sum

dict_sum(dict_a)
```
