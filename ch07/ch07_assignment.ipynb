```python
def check_odd_even(n):
  if n % 2 == 0:
    return "even"
  else:
    return "odd"


print(check_odd_even(4))
print(check_odd_even(7))
```

```python
def calculate_average(num_list:list):
  total = 0
  for i in num_list:
    total += i
  return total / len(num_list)

print("평균:", calculate_average([1, 2, 3, 4, 5]))
```
