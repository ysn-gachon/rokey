```python
import re
text = "이메일 목록: test@example.com, hello@world.net, user123@domain.org"

pattern = r"[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z]+"

emails = re.findall(pattern, text)
print(emails)
```

```python
import re
text = "연락처: 010-1234-5678, 02-987-6543, 031-456-7980"

pattern = r"[0-9-]+"

phone_numbers = re.findall(pattern, text)
print(phone_numbers)
```

```python
import re
text = "I love Python. Java is also popular. Python is great for AI."

pattern = r"\s*([^.]*Python[^.]*\.)"
py_sentences = re.findall(pattern, text)
print(py_sentences)
```

```python
import re
text = "상품 코드: A123, B456, C789, 가격: 12000원"

pattern = r"[0-9]+"
num = re.findall(pattern, text)
print(num)
```

```python
import re
text = "NASA is working on AI projects with IBM and Google"

pattern = r"[A-Z]{2,}"
capital = re.findall(pattern, text)
print(capital)
```
