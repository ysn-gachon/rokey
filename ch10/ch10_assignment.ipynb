```python
class Phone:
  def __init__(self, number, color):
    self.number = number
    self.color = color

phone = Phone("010-1234-5678", "검정")
print(phone.number)
print(phone.color)
```

```python
class SmartPhone(Phone):
  def __init__(self, number, color):
    super().__init__(number, color)

SmartPhone = SmartPhone("010-5678-5678", "그레이")
print(SmartPhone.number)
print(SmartPhone.color)
```

```python
class SmartPhone(Phone):
  def __init__(self, number, color, company):
    super().__init__(number, color)
    self.company = company

apple = SmartPhone("010-1234-5678", "검정", "애플")
print(apple.number)
print(apple.color)
print(apple.company)
```

```python
class Phone:
  def __init__(self, number, color):
    self.number = number
    self.color = color

  def showInfo(self):
    print("번호: ", self.number)
    print("색상: ", self.color)

phone = Phone("010-1234-5678", "검정")
phone.showInfo()
```

```python
class SmartPhone(Phone):
  def __init__(self, number, color, company):
    super().__init__(number, color)
    self.company = company

  def showInfo(self):
    super().showInfo()
    print("회사: ", self.company)

apple = SmartPhone("010-1234-5678", "검정", "애플")
apple.showInfo()
```
