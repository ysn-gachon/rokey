```python
numbers = [10, 20, 30]

try:
  index = int(input("인덱스를 입력해주세요: "))
  print(numbers[index])
except ValueError:
  print("숫자가 아닌 값이 입력되었습니다!")
except IndexError:
  print("잘못된 인덱스입니다.")
```
