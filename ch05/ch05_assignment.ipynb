```python
for i in range(1, 31):
  if i % 2 == 0:
    print(i, ": 짝수")
  else:
    print(i, ": 홀수")
```

```python
odd = []
even = []

for i in range(1, 31):
  if i % 2 == 0:
    even.append(i)
  else:
    odd.append(i)

print("홀수 리스트:", odd)
print("짝수 리스트:", even)
```
