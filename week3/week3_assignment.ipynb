```python
with open("data.txt", 'w', encoding='utf-8') as f:
  for i in range(1, 11):
    f.write(f"{i}번째 줄입니다.\n")

print("파일이 성공적으로 작성되었습니다.")

with open("data.txt", 'r', encoding='utf-8') as f:
  contents = f.read()

print("파일내용:")
print(contents)

with open("data.txt", 'a', encoding='utf-8') as f:
  f.write("11번째 줄입니다.\n")

print("내용이 성공적으로 추가되었습니다.")

with open("data.txt", 'r', encoding='utf-8') as f:
  contents = f.read()

print("파일내용:")
print(contents)
```

```python
while True:
  try:
    num = int(input("숫자를 입력해주세요: "))
    square = num*num
    print(square)
    break
  except ValueError:
    print("올바른 숫자를 입력하세요!")
```

```python
import re
data = """python one
life is too short
python two
you need python
python three"""

p = re.compile(r"python\s\w+$", re.M)
print(p.findall(data))
```

```python
import re
while True:
  try:
    email = input("이메일을 입력해주세요: ")
    text = "이메일 형식:" + email
    print(text)
    pattern = r"\w+@\w+\.\w{2,}"
    result = re.findall(pattern, text)
    if result == []:
      raise ValueError("이메일 형식이 올바르지 않습니다!")
    else:
      print(result)
      break
  except ValueError as e:
    print(e)
```
